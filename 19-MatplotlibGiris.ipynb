```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#Her seferinde plt.show() yazmamak için kullanılır
```

```python
yasListesi = np.array([20,30,22,11,37,65,19,71,6,12])
```

```python
kiloListesi = np.array([58,75,81,40,69,90,80,77,26,50])
```

```python
yasListesi
```

```python
kiloListesi
```

```python
plt.plot(yasListesi,kiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişimi")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
## 0-10 arası eşit aralıklarla 20 sayı üretilir
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi1
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "r--")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")
plt.subplot(1,2,2) # 1 sıra 2 kolon olacak, şu an 2.grafik çiziliyor
plt.plot(numpyDizisi2, numpyDizisi1, "r--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.6,0.6])
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.8,0.8])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2, numpyDizisi1, "b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Alt Grafik Başlık")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

#Birden fazla eksen olduğu için for loop kullanılmalı

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("X Ekseni")
    
plt.tight_layout() # iki grafik arasındaki mesafeyi açar
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, label = "numpyDizisi1 ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi1 ** 3")
yeniEksen.legend()
```

```python
yeniFigur.savefig("benimfigure.png", dpi=200)
#Kayıt
```
