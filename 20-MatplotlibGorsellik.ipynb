```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi = np.linspace(0,10,20)
```

```python
numpyDizisi
```

```python
(yeniFigur, yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi, numpyDizisi + 2, color="blue", linewidth=4)
yeniEksen.plot(numpyDizisi, numpyDizisi + 3, color="red")
yeniEksen.plot(numpyDizisi, numpyDizisi + 4, color="yellow", linestyle="--")
yeniEksen.plot(numpyDizisi, numpyDizisi + 5, color="black", linestyle="-.")
yeniEksen.plot(numpyDizisi, numpyDizisi + 6, color="purple", linestyle=":", linewidth=2)
##color değerine renk kodu da verilebilir
```

Scatter

```python
plt.scatter(numpyDizisi, numpyDizisi ** 2)
```
